Bit Error Rate (BER)

*  The Bit Error Rate (BER) is a measure of how many bits of data are incorrectly received compared to the total number of bits sent over a communication channel. It is expressed as a ratio or percentage.

*  The formula for calculating BER is:
  
*  BER = Number of Bit Errors / Total Number of Bits Sent


*  If you send 1,000 bits and 10 of those bits are received incorrectly, the BER would be:
*  BER = 10/1000 = 0.01 or 1%

Block Error Rate (BLER)
Definition:
The Block Error Rate (BLER) measures the number of blocks of data that are received incorrectly compared to the total number of blocks sent. A "block" typically refers to a group of bits that are transmitted together, often as part of a larger message.


- The formula for calculating BLER is:
 BLER = Number of Block Errors / Total Number of Blocks Sent



- If you send 100 blocks of data and 5 of those blocks are received incorrectly, the BLER would be:
BLER = 5/100 = 5%

```python
import torch
import torch.nn as nn
import numpy as np


class NeuralReceiver(nn.Module):
    def __init__(self):
        """
        Neural Receiver that processes received signals to remove noise and channel effects.
        """
        super(NeuralReceiver, self).__init__()

        # Calculate input dimensions based on y shape (128, 1, 1, 14, 76)
        self.input_features = 14 * 76  # 1064

        # Convolutional layers for feature extraction
        self.conv_layers = nn.Sequential(
            nn.Conv2d(1, 32, kernel_size=3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(),
        )

        # Calculate the size after convolutions
        self.conv_output_size = 128 * 14 * 76  # Maintaining spatial dimensions

        # Dense layers for signal processing
        self.fc_layers = nn.Sequential(
            nn.Linear(self.conv_output_size, 2048),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(2048, 1024),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(1024, 768),  # Output dimension matches required x shape
        )

    def forward(self, y, h_b, tau_b, h_out, no):
        """
        Forward pass of the neural receiver.

        Args:
            y (torch.Tensor): Received signal (128, 1, 1, 14, 76)
            h_b (torch.Tensor): Channel impulse response (128, 1, 1, 1, 16, 10, 1)
            tau_b (torch.Tensor): Delay profile (128, 1, 1, 10)
            h_out (torch.Tensor): Channel response (128, 1, 1, 1, 16, 1, 76)
            no (torch.Tensor): Noise power tensor of shape (1, 1)

        Returns:
            tuple: (x_hat, no_eff)
                - x_hat (torch.Tensor): Processed signal (128, 1, 1, 768)
                - no_eff (torch.Tensor): Effective noise power
        """
        batch_size = y.shape[0]
        # print("batch_size", batch_size)

        # Reshape input for 2D convolution: (batch, channels, height, width)
        y = y.view(batch_size, 1, 14, 76)
        # print("y shape", y.shape)

        # Apply convolution layers
        x = self.conv_layers(y)

        # Flatten for dense layers
        x = x.view(batch_size, -1)

        # Apply dense layers
        x = self.fc_layers(x)

        # Reshape output to match required dimensions (128, 1, 1, 768)
        x_hat = x.view(batch_size, 1, 1, -1)

        # Calculate effective noise power
        no_eff = self._calculate_effective_noise(x_hat, h_out, no)

        return x_hat, no_eff

    def _calculate_effective_noise(self, x_hat, h_out, no):
        """
        Calculate effective noise power based on the processed signal and channel response.

        Args:
            x_hat (torch.Tensor): Processed signal (batch_size, 1, 1, 768)
            h_out (torch.Tensor): Channel response (batch_size, 1, 1, 1, 16, 1, 76)
            no (torch.Tensor): Original noise power tensor of shape (1, 1)

        Returns:
            torch.Tensor: Effective noise power
        """
        # Calculate signal power while preserving batch dimension
        # Calculate signal power while preserving batch dimension
        signal_power = torch.mean(torch.abs(x_hat) ** 2)  # scalar

        # Calculate channel effect
        channel_effect = torch.mean(torch.abs(h_out) ** 2)  # scalar

        # Calculate effective noise power
        no_eff = no * signal_power * channel_effect

        # Ensure shape is (1, 1)
        no_eff = no_eff.view(1, 1)

        return no_eff  # Returns tensor of shape (1, 1)
```

```python
import torch
from torch.utils.data import DataLoader, TensorDataset

# Prepare the dataset
def prepare_data(dataset):
  tensor_dataset = []
  
  for index, (y, x, b, no, h_b, tau_b, h_out) in enumerate(dataset):
    # Convert numpy arrays to torch tensors
    y_torch = torch.from_numpy(y).float().requires_grad_(True)
    x_torch = torch.from_numpy(x).float().requires_grad_(True)
    b_torch = torch.from_numpy(b).float().requires_grad_(True)
    no_torch = torch.full((1, 1), no, dtype=torch.float32)
    h_b_torch = torch.from_numpy(h_b).float().requires_grad_(True)
    tau_b_torch = torch.from_numpy(tau_b).float().requires_grad_(True)
    h_out_torch = torch.from_numpy(h_out).float().requires_grad_(True)
    tensor_dataset.append((y_torch, x_torch, b_torch, no_torch, h_b_torch, tau_b_torch, h_out_torch))

  return tensor_dataset  

tensor_dataset = prepare_data(dataset)    
   
# dataloader = DataLoader(
#     dataset,
#     batch_size=1,
#     shuffle=True,
# )   
```

```python
import torch
import torch.optim as optim

# Initialize the model
model = NeuralReceiver()

# Define loss function and optimizer
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Number of epochs
num_epochs = 10

# Training loop
for epoch in range(num_epochs):
    total_loss = 0
    for y, x, b, no, h_b, tau_b, h_out in tensor_dataset:
        # Forward pass
        x_hat, no_eff = model(y, h_b, tau_b, h_out, no)

        # Compute loss
        loss = criterion(x_hat, x)

        # Backward pass and optimize
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    # Print average loss for the epoch
    avg_loss = total_loss / len(tensor_dataset)
    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {avg_loss:.4f}")
```

```python
# Evaluation
model.eval()
with torch.no_grad():
    total_ber = 0
    num_samples = 0
    for y, x, b, no, h_b, tau_b, h_out in tensor_dataset:
        x_hat, no_eff = model(y, h_b, tau_b, h_out, no)

        # Use the provided evaluation functions
        b_hat, _ = eval_transceiver.demapper_decision(
            x_hat=x_hat.numpy(), no_eff=no_eff.item()
        )
        ber = calculate_BER(b.numpy(), b_hat)

        total_ber += ber * len(y)
        num_samples += len(y)

    average_ber = total_ber / num_samples
    print(f"Average BER: {average_ber:.6f}")
```

```python
# Initialize model, loss function, and optimizer
model = NeuralReceiver()
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Set the target BER
target_ber = 0.1

# Training loop
epoch = 0
max_epochs = 1000  # Set a maximum number of epochs to prevent infinite loops
while True:
    epoch += 1
    total_loss = 0

    for y, x, b, no, h_b, tau_b, h_out in tensor_dataset:
        # Forward pass
        x_hat, no_eff = model(y, h_b, tau_b, h_out, no)

        # Compute loss
        loss = criterion(x_hat, x)

        # Backward pass and optimize
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    # Print average loss for the epoch
    avg_loss = total_loss / len(tensor_dataset)
    print(f"Epoch [{epoch}/{max_epochs}], Loss: {avg_loss:.4f}")

    # Evaluate BER
    model.eval()
    with torch.no_grad():
        total_ber = 0
        num_samples = 0
        for y, x, b, no, h_b, tau_b, h_out in tensor_dataset:
            x_hat, no_eff = model(y, h_b, tau_b, h_out, no)

            # Convert to numpy for evaluation
            x_hat_np = x_hat.numpy()
            no_eff_np = no_eff.item()

            # Use the provided evaluation functions
            b_hat, _ = eval_transceiver.demapper_decision(
                x_hat=x_hat_np, no_eff=no_eff_np
            )
            ber = calculate_BER(b.numpy(), b_hat)

            total_ber += ber * len(y)
            num_samples += len(y)

        average_ber = total_ber / num_samples
        print(f"Average BER: {average_ber:.6f}")

    # Check if target BER is achieved
    if average_ber < target_ber:
        print(f"Target BER of {target_ber} achieved. Training complete.")
        break

    # Check if maximum epochs reached
    if epoch >= max_epochs:
        print(f"Maximum epochs ({max_epochs}) reached. Training stopped.")
        break

    model.train()  # Set model back to training mode

print(f"Final Average BER: {average_ber:.6f}")
```
